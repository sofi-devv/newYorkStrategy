Tenemos demasiadas categorías, esto podría generar problemas a la hora de implementar un modelo, por eso se normalizan en categorías más sencillas y compactas. Nuevamente con la intención de tener una comprensión general del estado de nueva york en estos problemas.

## Categorías de Delitos:

1. **Delitos contra la propiedad**:
   - Descripción: Esta categoría incluye delitos que implican el robo, el hurto, la intrusión ilegal en propiedad ajena, el daño a la propiedad, entre otros.
   
2. **Delitos contra las personas**:
   - Descripción: Esta categoría abarca delitos que afectan directamente a la integridad física o psicológica de las personas, como agresiones, homicidios, secuestros, etc.
   
3. **Delitos contra la seguridad pública**:
   - Descripción: Engloba delitos relacionados con el consumo y tráfico de drogas, alcohol, prostitución, juego ilegal, conducción bajo la influencia, entre otros, que representan un riesgo para la seguridad y el orden público.
   
4. **Delitos contra la administración pública**:
   - Descripción: Incluye delitos que afectan el funcionamiento de las instituciones y servicios públicos, como el uso ilegal de armas, violaciones a normativas de orden público, etc.
   
5. **Otros delitos**:
   - Descripción: Esta categoría agrupa delitos que no encajan claramente en las categorías anteriores, como delitos menores, delitos relacionados con el tráfico vehicular, fraudes, delitos contra la infancia, etc. También incluye delitos que pueden ser considerados inusuales o peculiares.


```python
# Define el diccionario de mapeo para las cinco categorías
category_map = {
    'Delitos contra la propiedad': ['ROBBERY', 'ARSON', 'GRAND LARCENY OF MOTOR VEHICLE', 'PETIT LARCENY', 'GRAND LARCENY', 'BURGLARY', 'CRIMINAL MISCHIEF & RELATED OF', 'POSSESSION OF STOLEN PROPERTY', 'BURGLAR\'S TOOLS', 'THEFT OF SERVICES', 'THEFT-FRAUD'],
    'Delitos contra las personas': ['ASSAULT 3 & RELATED OFFENSES', 'FELONY ASSAULT', 'JOSTLING', 'SEX CRIMES', 'MURDER & NON-NEGL. MANSLAUGHTE', 'HOMICIDE-NEGLIGENT,UNCLASSIFIE', 'KIDNAPPING & RELATED OFFENSES', 'CHILD ABANDONMENT/NON SUPPORT', 'HARRASSMENT 2'],
    'Delitos contra la seguridad pública': ['DANGEROUS DRUGS', 'CANNABIS RELATED OFFENSES', 'ALCOHOLIC BEVERAGE CONTROL LAW', 'GAMBLING', 'PROSTITUTION & RELATED OFFENSES', 'INTOXICATED & IMPAIRED DRIVING', 'INTOXICATED/IMPAIRED DRIVING'],
    'Delitos contra la administración pública': ['DANGEROUS WEAPONS', 'OFF. AGNST PUB ORD SENSBLTY &', 'OFFENSES AGAINST PUBLIC ADMINI', 'OFFENSES AGAINST PUBLIC SAFETY', 'ADMINISTRATIVE CODE', 'ADMINISTRATIVE CODES'],
    'Otros delitos': ['MISCELLANEOUS PENAL LAW', 'VEHICLE AND TRAFFIC LAWS', 'OTHER OFFENSES RELATED TO THEF', 'OFFENSES INVOLVING FRAUD', 'OFFENSES RELATED TO CHILDREN', 'DISRUPTION OF A RELIGIOUS SERV', 'ENDAN WELFARE INCOMP', 'OTHER STATE LAWS', 'OTHER STATE LAWS (NON PENAL LA', 'OTHER STATE LAWS (NON PENAL LAW)', 'ANTICIPATORY OFFENSES', 'FORTUNE TELLING']
}

# Función para obtener la categoría de un delito
def get_category(ofns_desc):
    for category, ofns_list in category_map.items():
        if ofns_desc in ofns_list:
            return category
    return 'No categorizado'

# Aplicar el mapeo a la columna 'OFNS_DESC'
arrests['NEW_CATEGORY'] = arrests['OFNS_DESC'].apply(get_category)

# Verificar los valores únicos después del mapeo
print(arrests['NEW_CATEGORY'].unique())
```
